## Data Science - Task-1

#### To write a script to scrape ads data from a given url along with suitable information and store Image url, Model name, price and other information to a csv/excel file.

```python
# importing required libraries

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# target URL to scrap

url = "https://www.google.com/search?q=samsung+mobile&rlz=1C1YTUH_enIN947IN950&oq=samsung+mobile&aqs=chrome.0.69i59l2j69i60l3.8727j0j7&sourceid=chrome&ie=UTF-8"

headers = {
    'User-Agent': "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36"
    }
```

```python
# send request to download the data

response = requests.request("GET", url, headers=headers)
```

```python
# parse the downloaded data

data = BeautifulSoup(response.text, 'html.parser')
print(data)
```

```python
# find all the sections with specifiedd class name

mobiles_data = data.find_all('div', attrs={'class': "mnr-c pla-unit"})
```

```python
print(mobiles_data)
```

```python
# total number of mobiles
print('Total Number of Mobiles Found : ', len(mobiles_data))
```

```python
# extract the mobile phones, model name and price
for mobile in mobiles_data:

    img_url = mobile.find_all('img')
    model_name = mobile.find('span', attrs={'class' : "pymv4e"}).text
    model_price = mobile.find('div', attrs={'class' : "qptdjc"}).text
    model_seller = mobile.find('span', attrs={'class' : "zPEcBd LnPkof"}).text

    print(img_url, model_name, model_price, model_seller)
```

```python
# create a list to store the data

scraped_data = []
```

```python
for mobile in mobiles_data:

    # initialize the dictionary
    mobile_details = {}
    
    img_url = mobile.find_all('img')
    model_name = mobile.find('span', attrs={'class': "pymv4e"}).text
    model_price = mobile.find('div', attrs={'class': "qptdjc"}).text
    model_seller = mobile.find('span', attrs={'class': "zPEcBd LnPkof"}).text
    
     # add data to the dictionary
    mobile_details['img_url'] = img_url
    mobile_details['model_name'] = model_name
    mobile_details['model_price'] = model_price
    mobile_details['model_seller'] = model_seller
    
    # append the scraped data to the list
    scraped_data.append(mobile_details)
```

```python
print(scraped_data)
```

```python
# create a data frame from the list of dictionaries
dataFrame = pd.DataFrame.from_dict(scraped_data)

# save the scraped data as CSV file
dataFrame.to_csv('mobiles_data.csv', index=False)
```
